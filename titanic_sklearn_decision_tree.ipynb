```
import numpy as np
import pandas as pd
from IPython.display import display # Allows the use of display() for DataFrames

# Pretty display for notebooks
%matplotlib inline

# Set a random seed
import random
random.seed(42)

# Load the dataset
in_file = 'data.csv'
full_data = pd.read_csv(in_file)

# Print the first few entries of the RMS Titanic data
display(full_data.head())
```

```
# Store the 'Survived' feature in a new variable and remove it from the dataset
outcomes = full_data['Survived']
features_raw = full_data.drop('Survived', axis = 1)

# Show the new dataset with 'Survived' removed
display(features_raw.head())
```

```
# Removing the names
features_no_names = features_raw.drop(['Name'], axis=1)

# One-hot encoding
features = pd.get_dummies(features_no_names)
display(features.head())
```

```
features = features.fillna(0.0)
display(features.head())
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state=42)
```

```
from sklearn.tree import DecisionTreeClassifier

# TODO: Define the classifier, and fit it to the data
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```

```
#Using Grid Search to find the optimal value for max_depth, min_smaples_leaf, min_samples_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score

scorer = make_scorer(f1_score)

parameters = {'max_depth': range(1,30), 'min_samples_leaf':range(1,30), 'min_samples_split': range(2,30)}

dmodel = DecisionTreeClassifier(random_state=42)

grid = GridSearchCV(dmodel,parameters,scorer, cv=4)

grid = grid.fit(X_train, y_train)
best_clf = grid.best_estimator_
```

```
print(best_clf.get_params()['max_depth'])
print(best_clf.get_params()['min_samples_leaf'])
print(best_clf.get_params()['min_samples_split'])
```

```
# TODO: Make predictions
dy_train_pred = best_clf.predict(X_train)
dy_test_pred = best_clf.predict(X_test)

# TODO: Calculate the accuracy
dtrain_accuracy = accuracy_score(y_train, dy_train_pred)
dtest_accuracy = accuracy_score(y_test, dy_test_pred)
print('The training accuracy is', dtrain_accuracy)
print('The test accuracy is', dtest_accuracy)
```
